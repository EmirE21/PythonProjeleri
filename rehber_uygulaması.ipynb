```python
# Kişilerin Depolanacağı Boş Sözlük
guide_dict = {}
# Menü
def menu():
  print("Rehber Uygulamasına Hoşgeldin!")
  print("1. Kişi Ekle")
  print("2. Kişi Sil")
  print("3. Kişi Güncelle")
  print("4. Kişi Ara")
  print("5. Kişileri Listele")
  print("6. Çıkış")
# Kişi Ekleme
def add_person(guide_dict):
  name = input("Kişi Adı ve Soyadı (Aralarına boşluk koyun ve büyük harfle başlayın lütfen): ")
  phone = input("Kişi Telefon Numarası: ")
  while True:
    email = input("Kişi Email Adresi: ").lower()
    if "@" and "." not in email:
      print("Hatalı email girşi!")
      continue
    else:
      break
  guide_dict[name] = {
      "Phone": phone,
      "Email": email
  }
  print("Kişi Başarıyla Eklendi.")
# Kişi Silme
def delete_person(guide_dict):
  while True:
    name = input("Silmek istediğiniz kişinin adı ve soyadı: ")
    if name in guide_dict:
      del guide_dict[name]
      break
    else:
      print("Kişi bulunamadı! Lütfen tekrar deneyin.")
      continue
  print("Kişi başarıyla silindi.")
# Kişi Güncelleme
def update_person(guide_dict):
  while True:
    name = input("Güncellemek istediğiniz kişinin adı ve soyadı:")
    break
    if name not in guide_dict.keys():
      print("Kişi bulunamadı! Lütfen tekrar deneyin.")
      continue

  while True:
    if name in guide_dict.keys():
      infos = input("Değiştirmek istediğiniz kısımları girin (Çoklu seçim yapacaksanız virgülle ayırın lütfen): ").lower().split(",")
      if "phone" in infos:
        new_phone = input("Yeni telefon numarası: ")
        guide_dict[name]["Phone"] = new_phone
        break
      elif "email" in infos:
        new_email = input("Yeni email adresi: ")
        guide_dict[name]["Email"] = new_email
        break
      elif "phone" and "email" in infos:
        new_phone = input("Yeni telefon numarası: ")
        new_email = input("Yeni email adresi: ")
        guide_dict[name]["Phone"] = new_phone
        guide_dict[name]["Email"] = new_email
        break
      else:
        print("Hatalı giriş! Lütfen tekrar deneyin.")
        continue
  print("Kişi başarıyla güncellendi.")
# Tek Bir Kişi Aratma
def search_person(guidce_dict):
  while True:
    name = input("Aramak istediğiniz kişinin adı ve soyadı: ")
    if name in guide_dict:
      return print(guide_dict[name])
    else:
      print("Hatalı giriş! Lütfen tekrar deneyin.")
# İşlem Döngüsü
menu()
while True:
  choice = input("Seçiminiz: ")
  if choice == "1":
    add_person(guide_dict)
  elif choice == "2":
    delete_person(guide_dict)
  elif choice == "3":
    update_person(guide_dict)
  elif choice == "4":
    search_person(guide_dict)
  elif choice == "5":
    print(guide_dict)
  elif choice == "6":
    print("Başarıyla çıkış yapılmıştır.")
    break
  else:
    print("Hatalı giriş! Lütfen tekrar deneyin.")
```
