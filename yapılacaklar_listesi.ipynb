```python
to_do_list = {}
# Menü
def menu():
  print("= Yapılacaklar Listesi =")
  print("1. Madde Ekle")
  print("2. Madde Sil")
  print("3. Madde Değiştir")
  print("4. Listeyi Göster")
  print("5. Çıkış")
# Ekle
def add(to_do_list):
  while True:
    try:
      num = int(input("Yazılacak işin sıra numarasını belirtin lütfen: "))
      break
    except ValueError:
      print("Geçersiz giriş. Lütfen bir sayı girin.")
      continue

  job = input("Yapacağınız işi girin: ")
  time = input("İşinizi ne zaman yapacağınızı girin: ")
  status = input("İşin yapılıp yapılmadığını girin: ")

  to_do_list[num] = {
      "Yapılacak iş": job,
      "Yapılacak zaman": time,
      "Yapılma durumu": status
  }

  while True:
    question= input("Başka bir iş eklemek istiyor musunuz? (Evet/Hayır): ").lower()
    if question == "evet":
      add(to_do_list)
      break
    elif question == "hayır":
      break
    else:
      print("Yanlış değer girdiniz, lütfen tekrar deneyin.")
      continue
# Sil
def delete(to_do_list):
  while True:
    try:
      num = int(input("Silmek istediğiniz işinizin numarasını girin: "))
      if num in to_do_list:
        del to_do_list[num]
        print(f"Madde {num} silindi.")
        break
      else:
        print(f"Madde {num} listede bulunamadı.")
    except ValueError:
      print("Geçersiz giriş. Lütfen bir sayı girin.")

# Değiştir
def change(to_do_list):
  while True:
    try:
      num = int(input("Değiştirmek istediğiniz işinizin numarasını girin: "))
      if num in to_do_list:
        break
      else:
        print(f"Madde {num} listede bulunamadı.")
        continue
    except ValueError:
      print("Geçersiz giriş. Lütfen bir sayı girin.")

  while True:
    text = input("Değiştirmek istediğiniz kısmı girin (Yapılacak iş/Yapılacak zaman/Yapılma durumu): ").lower()
    if text == "yapılacak iş":
      to_do_list[num]["Yapılacak iş"] = input("Yeni yapılacak işi girin: ")
      break
    elif text == "yapılacak zaman":
      to_do_list[num]["Yapılacak zaman"] = input("Yeni yapılacak zamanı girin: ")
      break
    elif text == "yapılma durumu":
      to_do_list[num]["Yapılma durumu"] = input("Yeni yapılma durumu girin: ")
      break
    else:
      print("Yanlış değer girdiniz, lütfen tekrar deneyin.")
      continue

  while True:
    question= input("Başka bir değişiklik yapmak istiyor musunuz? (Evet/Hayır)").lower()
    if question == "evet":
      change(to_do_list)
      break
    elif question == "hayır":
      break
    else:
      print("Yanlış değer girdiniz, lütfen tekrar deneyin.")
      continue
# Listeyi Göster
def show_list(to_do_list):
  if to_do_list:
    for num, details in to_do_list.items():
      print(f"{num}:")
      for key, value in details.items():
        print(f"{key}: {value}")
  else:
    print("Yapılacaklar listesi boş.")

# işlem Döngüsü
menu()
while True:
  choice = input("Lütfen yapmak istediğiniz işlemin numarasını girin: ")
  if choice == "1":
    add(to_do_list)
  elif choice == "2":
    delete(to_do_list)
  elif choice == "3":
    change(to_do_list)
  elif choice == "4":
    show_list(to_do_list)
  elif choice == "5":
    break
  else:
    print("Geçersiz seçim. Lütfen 1 ile 5 arasında bir sayı girin.")
```
