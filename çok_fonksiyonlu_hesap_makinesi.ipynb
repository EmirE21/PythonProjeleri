```python
# Matematik Modülü
from math import factorial, sin, cos, tan, log, log10, asin, acos, atan, pi, degrees, e
# Menü
def operation_list():
  print("= Çok Fonksiyonlu Hesap Makinesi=")
  print("0. Çıkış")
  print("1. Toplama")
  print("2. Çıkarma")
  print("3. Çarpma")
  print("4. Bölme")
  print("5. Üs Alma")
  print("6. Kök Alma")
  print("7. Faktöriyel")
  print("8. Logaritma")
  print("9. Doğal Logaritma")
  print("10. Sinüs")
  print("11. Kosinüs")
  print("12. Tanjant")
# Toplama
def addition():
  term_number = int(input("Toplamak istediğiniz sayı miktarını girin: "))
  print("Not: Lütfen ondalıklı sayılarda '.' kullanın")

  total = 0
  for i in range(term_number):
    number = input(f"{i+1}. sayıyı girin: ")
    if "." in number:
      total += float(number)
    else:
      total += int(number)
  print(f"Sonuç: {total}")
# Çıkarma
def subtraction():
  term_number = int(input("Çıkarmak istediğiniz sayı miktarını girin: "))
  print("Not: Lütfen ondalıklı sayılarda '.' kullanın")

  total = 0
  # İlk sayı pozitif yapılır.
  first_number = input("1. sayıyı girin: ")
  if "." in first_number:
    total += float(first_number)
  else:
    total += int(first_number)

  for i in range(term_number-1):
    number = input(f"{i+2}. sayıyı girin: ")
    if "." in number:
      total -= float(number)
    else:
      total -= int(number)
  print(f"Sonuç: {total}")
# Çarpma
def multiplication():
  term_number = int(input("Çarpmak istediğiniz sayı miktarını girin: "))
  print("Not: Lütfen ondalıklı sayılarda '.' kullanın")

  total = 1
  for i in range(term_number):
    number = input(f"{i+1}. sayıyı girin: ")
    if "." in number:
      total *= float(number)
    else:
      total *= int(number)
  print(f"Sonuç: {total}")
# Bölme
def division():
  while True:
    num1 = input("1. sayıyı girin: ")
    num2 = input("2. sayıyı girin: ")
    try:
      if num2 == 0:
        raise ZeroDivisionError("Payda sıfır olamaz.")
      if "." in num1 or "." in num2:
        result = float(num1) / float(num2)
      elif int(num1) < int(num2):
        result = int(num1) / int(num2)
      else:
        result = int(num1) // int(num2)
      print(f"Sonuç: {result}")
      break
    except ZeroDivisionError as zde:
      print(f"Uyarı: {zde}")
      continue
# Üslü Sayı
def exponential():
  print("Lütfen üs ve taban kısmına kesirli bir sayı girmeyin.")
  while True:
    exponent = int(input("Lütfen almak istediğiniz üssü girin: "))
    base = int(input("Lütfen üssü alınacak sayıyı girin: "))
    try:
      if exponent < 0 and base == 0:
        raise ZeroDivisionError("Üs negatif iken taban değer sıfır olamaz.")
      elif base < 0:
        result1 = base ** exponent
        result2 = -base ** exponent
        print(f"Üs parantez dışında olursa sonuç: {result1}")
        print(f"Üs parantez içinde olursa sonuç: {result2}")
      else:
        result = base ** exponent
        print(f"Sonuç: {result}")
        break
    except ZeroDivisionError as zde:
      print(f"Uyarı: {zde}")
      continue
# Köklü Sayı
def root():
  while True:
    root_degree = int(input("Lütfen kökün derecesini girin: "))
    num = int(input("Lütfen kök alınacak sayıyı girin: "))
    try:
      if root_degree < 0:
        raise ValueError("Kök derecesi negatif olamaz.")
      elif root_degree == 2 and num < 0:
        raise ValueError("Karekökü alınacak sayı negatif olamaz.")
      elif root_degree == 0 and num == 0:
        raise ZeroDivisionError("Sıfırın sıfır kökü alınamaz.")
      else:
        print(f"Sonuç: {num ** (1 / root_degree)}")
        break
    except ValueError as ve:
      print(f"Uyarı: {ve}")
      continue
    except ZeroDivisionError as zde:
      print(f"Uyarı: {zde}")
      continue
# Logaritma
def logarithm():
  while True:
    print("Logaritma içerisinde rasyonel ifade olmasını istemiyorsanız ikinci sayıyı '1' olarak girin.")
    base = int(input("Logaritma tabanını girin: "))
    num1 = float(input("1. sayıyı girin: "))
    num2 = float(input("2. sayıyı girin: "))
    try:
      if base == 0:
        raise ValueError("Logaritmanın tabanı sıfır olamaz.")
      elif base < 0:
        raise ValueError("Logaritmanın tabanı negatif olamaz.")
      elif base == 1:
        raise ValueError(" Logaritmanın tabanı 1 olamaz.")
      elif num1 / num2 == 0:
        raise ValueError("Logaritması alınan sayı sıfıra eşit olamaz.")
      elif num1 / num2 < 0:
        raise ValueError("Logaritması alınan sayı negatif olamaz.")
      elif num2 == 0:
        raise ZeroDivisionError("Logaritması alınan kesirli sayının tabanı sıfır olamaz.")
      if base == 10:
        result = log10(num1 / num2)
      else:
        result = log(num1 / num2, base)
      print(f"Sonuç: {result}")
      break
    except ZeroDivisionError as zde:
      print(f"Uyarı: {zde}")
      continue
    except ValueError as ve:
      print(f"Uyarı: {ve}")
      continue
# Doğal Logaritma
def natural_logarithm():
  while True:
    num = input("Logaritması alınacak sayıyı girin: ")
    try:
      if num == "e":
        result = e
      elif float(num) < 0:
        raise ValueError("Logaritması alınan sayı negatif olamaz.")
      else:
        result = log(float(num))
      print(f"Sonuç: {result}")
      break
    except ValueError as ve:
      print(f"Uyarı: {ve}")
      continue
# Faktöriyel
def factorial_process():
  while True:
    num = int(input("Lütfen bir tam sayı girin: "))
    try:
      if num < 0:
        raise ValueError("Faktöriyel değeri negatif olamaz.")
      else:
        print(f"Sonuç: {factorial(num)}")
        break
    except ValueError as ve:
      print(f"Uyarı: {ve}")
      continue
# Sinüs
def sinus():
  print("Not: Lütfen radyan cinsinden açı girmeyin. Tam sayı olan bir değer girin.")
  sin_angel = float(input("Sinüs değerini hesaplamak istediğiniz açıyı girin: "))
  sin_angel *= pi / 180
  result = sin(sin_angel)
  print(f"Sonuç: {result}")
  while True:
    sin_angel = float(input("Ters sinüs değerini hesaplamak istediğiniz açıyı girin: "))
    try:
        if not -1 <= sin_angel <= 1:
          raise ValueError("Ters sinüs değeri -1 ile 1 arasında olmalıdır.")
        else:
          result = degrees(asin(sin_angel))
          print(f"Sonuç: {result}")
          break
    except ValueError as ve:
      print(f"Uyarı: {ve}")
      continue
# Kosinüs
def cosine():
  print("Not: Lütfen radyan cinsinden açı girmeyin. Tam sayı olan bir değer girin.")
  cos_angel = int(input("Kosinüs değerini hesaaplamak istediğiniz açıyı girinÇ: "))
  cos_angel *= pi / 180
  result = cos(cos_angel)
  print(f"Sonuç: {result}")
  while True:
    cos_angel = float(input("Ters kosinüs değerini hesaplamak istediğiniz açıyı girin: "))
    try:
      if not -1 <= cos_angel <= 1:
        raise ValueError("Ters kosinüs değeri -1 ile 1 arasında olmalıdır.")
      else:
        result = degrees(acos(cos_angel))
        print(f"Sonuç: {result}")
        break
    except ValueError as ve:
      print(f"Uyarı: {ve}")
      continue
# Tanjant
def tangent():
  while True:
    tan_angel = int(input("Tanjant değerini hesaplamak istediğiniz bir açıyı girin: "))
    tan_angel *= pi / 180
    result = tan(tan_angel)
    try:
      if tan_angel == 90 and (tan_angel / 90) % 2 != 0:
        raise ValueError("Tanjant değeri 90 ve 180 fazlası olan katları olamaz.")
      else:
        print(f"Sonuç: {result}")
        break
    except ValueError as ve:
      print(f"Uyarı: {ve}")
      continue
  tan_angel = float(input("Ters tanjant değerini hesaplamak istediğiniz bir açıyı girin: "))
  result = degrees(atan(tan_angel))
  print(f"Sonuç: {result}")
# İşlem Gerçekleştirme Döngüsü
operation_list()
while True:
  choice = input("Lütfen yapmak istediğiniz işlem numarasını girin: ")
  if choice == "1":
    addition()
  elif choice == "2":
    subtraction()
  elif choice == "3":
    multiplication()
  elif choice == "4":
    division()
  elif choice == "5":
    exponential()
  elif choice == "6":
    root()
  elif choice == "7":
    factorial_process()
  elif choice == "8":
    logarithm()
  elif choice == "9":
    natural_logarithm()
  elif choice == "10":
    sinus()
  elif choice == "11":
    cosine()
  elif choice == "12":
    tangent()
  elif choice == "0":
    print("Hesap makinesinden ıkış yapıldı.")
    break
  else:
    print("Lütfen geçerli bir işlem numarası girin.")
```
